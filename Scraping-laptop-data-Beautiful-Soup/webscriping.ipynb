```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
import io
```

```python
k=1
b=0
c=0
```

```python
while k <=50:
    headers={'User-Agent':'#useragent'}
    r=requests.get("https://www.hepsiburada.com/ara?q=laptop&kategori=2147483646_3000500_98&sayfa="+str(k)+"", headers=headers)
    soup=BeautifulSoup (r.content, "lxml")
    st1=soup.find("div", attrs={"class":"productListContent-pXUkO4iHa51o_17CBibU"})
    print(k)
    print(b)
    while c <= 27:
        c=c+1
        st2 = st1.find("ul", attrs={"id": k})

        if st2 is not None:
            st3 = st2.find("li", attrs={"id": "i" + str(b) + ""})
            b=b+1
            if st3 is not None:
                for icerik in st3:
                    linksonu = icerik.a.get("href")
                    linkbasi = "https://www.hepsiburada.com/"

                    if not linksonu.startswith("https://adservice.hepsiburada.com/"):
                        link = linkbasi + linksonu
                    else:
                        link = linksonu

                    print(link)

                    headers = {
                        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'
                    }

                    r1 = requests.get(link, headers=headers)
                    soup1 = BeautifulSoup(r1.content, "lxml")

                    ozellikler = soup1.find("table", attrs={"class": "data-list tech-spec"})
                    fiyatlar = soup1.find("span", attrs={"data-bind": "markupText:'currentPriceBeforePoint'"}).text
                    marka = soup1.find("a", attrs={
                        "data-bind": "attr: {'data-hbus': userInformation() && userInformation().userId && isEventReady() ? productDetailHbus('BrandClick') : '' }"}).text

                    marka1 = {'Marka': marka}
                    fiyatlar1 = {'Fiyat': fiyatlar}
                    marka1.update(fiyatlar1)

                    degiskenler = ['Bellek Hızı', 'Bluetooth Özelliği', 'Cihaz Ağırlığı', 'Dokunmatik Ekran',
                                   'Ekran Boyutu', 'Ekran Kartı Bellek Tipi', 'Ekran Kartı Hafızası', 'Ekran Kartı Tipi',
                                   'Ekran Kartı', 'Ekran Panel Tipi', 'eMMC Kapasitesi', 'Harddisk Kapasitesi', 'HDMI',
                                   'İşlemci Cache', 'İşlemci Nesli', 'İşlemci Tipi', 'İşlemci', 'İşletim Sistemi',
                                   'Kart Okuyucu', 'Klavye', 'Maksimum İşlemci Hızı', 'Max Ekran Çözünürlüğü',
                                   'Optik Sürücü', 'Parmak İzi Okuyucu', 'Ram (Sistem Belleği)', 'Ram Tipi', 'Renk',
                                   'SSD Kapasitesi', 'Ürün Modeli', 'Webcam']

                    trler = ozellikler.find_all('tr')
                    degerler = {}

                    for degisken in degiskenler:
                        for tr in trler:
                            th = tr.find('th')
                            if th and th.text.strip() == degisken:
                                td = tr.find('td')
                                if td:
                                    deger = td.text.strip()
                                    degerler[degisken] = deger
                                    break

                    print(degerler)
                    degerler.update(marka1)

                    dosya_adi = 'deneme3.csv'

                    with open(dosya_adi, 'a', newline='',encoding="utf-8") as csvfile:
                        writer = csv.writer(csvfile)
                        veri = list(degerler.values())
                        writer.writerow(veri)
            else:
                print("st3 bulunamadı")
        else:
            print("st2 bulunamadı")

    k = k + 1
    b = b - 4
    c = 0
```
